```python
import os
H2O = -44.03599425
ZINC =  -22155.05314789
H2O_ZINC = H2O + ZINC
```

```python
import os
import pandas as pd 
import matplotlib.pyplot as plt 

# CSV  
csv_files = [x for x in os.listdir() if ".csv" in x]
csv_file = csv_files[0]

#  
df = pd.read_csv(csv_file)
energy = df["energy"].to_list()

#    (H2O_ZINC  )
ad_energy_kj = [ (e - H2O_ZINC) * 2625.49884478051 for e in energy]
ad_energy_ev = [ (e - H2O_ZINC) * 27.211324570273 for e in energy]

plt.figure(figsize=(10, 6))
#  5    (alpha=0.6)  scatter plot
plt.scatter(range(len(ad_energy_kj[5:])), ad_energy_kj[5:], alpha=0.6, label="H2O")
plt.xlabel("Iteration Step")
plt.ylabel("Total Energy (kJ)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
